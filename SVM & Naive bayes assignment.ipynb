### SVM & Naive Bayes: Theoretical Assignment Questions/Answers

**1. What is a Support Vector Machine (SVM)?**
     
-->
A Support Vector Machine (SVM) is a supervised machine learning algorithm used for both classification and regression tasks. [cite_start]Its primary goal is to find an optimal hyperplane that best separates data points belonging to different classes in a high-dimensional space[cite: 2].

**2. What is the difference between Hard Margin and Soft Margin SVM?**

-->
* **Hard Margin SVM:** This type of SVM seeks a hyperplane that perfectly separates the classes without allowing any misclassifications. [cite_start]It is highly sensitive to outliers and is only applicable when the data is linearly separable[cite: 2].

* **Soft Margin SVM:** In contrast, Soft Margin SVM permits some misclassifications to achieve a more robust model. The degree of allowed misclassification is controlled by the 'C' parameter. [cite_start]This approach is more practical for real-world datasets that are often not perfectly linearly separable[cite: 2].

**3. What is the mathematical intuition behind SVM?**

-->

The core mathematical intuition behind SVM is to identify a hyperplane that maximizes the margin—the distance between the hyperplane and the closest data points from each class. These closest data points are known as support vectors. [cite_start]The underlying optimization problem involves minimizing a cost function, which incorporates both the margin size and, for Soft Margin SVM, the penalty for misclassification errors[cite: 3].

**4. What is the role of Lagrange Multipliers in SVM?**

-->

Lagrange Multipliers are crucial in SVM for solving its constrained optimization problem. [cite_start]They help transform the original problem into its dual form, which is often simpler to solve and facilitates the application of the kernel trick[cite: 4].

**5. What are Support Vectors in SVM?**

-->

Support vectors are the data points from the training set that lie nearest to the decision boundary (hyperplane). These specific data points are critical because they define the position and orientation of the optimal hyperplane. [cite_start]Removing a support vector would typically alter the hyperplane's position[cite: 5].

**6. What is a Support Vector Classifier (SVC)?**

-->

A Support Vector Classifier (SVC) is the classification variant of SVM. [cite_start]It is used when the target variable is discrete, and the objective is to categorize data into distinct classes[cite: 6].

**7. What is a Support Vector Regressor (SVR)?**

-->

A Support Vector Regressor (SVR) is the regression variant of SVM. Unlike SVC, SVR predicts continuous output values. [cite_start]It aims to find a hyperplane that best fits the data points within a specified margin of tolerance, rather than classifying them[cite: 7].

**8. What is the Kernel Trick in SVM?**

-->

The Kernel Trick is a powerful technique in SVM that allows it to operate effectively in a high-dimensional, implicit feature space without explicitly calculating the coordinates of the data in that space. It achieves this by using kernel functions (such as polynomial or Radial Basis Function - RBF) to compute the dot product between transformed feature vectors. [cite_start]This enables SVM to handle non-linearly separable data by implicitly mapping it to a higher dimension where it becomes linearly separable[cite: 8].

**9. Compare Linear Kernel, Polynomial Kernel, and RBF Kernel.**

-->

* [cite_start]**Linear Kernel:** This kernel is best suited for datasets that are linearly separable, meaning the decision boundary is a straight line or a simple hyperplane[cite: 9].
* **Polynomial Kernel:** Designed for non-linearly separable data, the Polynomial Kernel maps the data into a higher-dimensional space using polynomial functions. [cite_start]It requires parameters such as the degree and a constant term[cite: 9].
* **RBF (Radial Basis Function) Kernel:** A widely used and highly versatile kernel, the RBF Kernel is excellent for non-linear data. [cite_start]It creates complex, non-linear decision boundaries and has a `gamma` parameter that controls the "reach" or influence of individual training examples[cite: 9].

**10. What is the effect of the C parameter in SVM?**

--.

[cite_start]The `C` parameter in SVM manages the trade-off between maximizing the margin and minimizing the misclassification errors[cite: 10].
* A **small C value** results in a wider margin but allows for more misclassifications, potentially leading to underfitting.
* [cite_start]A **large C value** creates a narrower margin but strives for fewer misclassifications, which can sometimes lead to overfitting[cite: 10].

**11. What is the role of the Gamma parameter in RBF Kernel SVM?**

-->

[cite_start]The `gamma` parameter in the RBF Kernel determines the influence of a single training example on the decision boundary[cite: 11].
* A **small gamma** implies a large influence, leading to a smoother, more generalized decision boundary (which might underfit).
* [cite_start]A **large gamma** implies a small influence, resulting in a more complex, potentially overfitted decision boundary that tries to fit every training example closely[cite: 11].

**12. What is the Naïve Bayes classifier, and why is it called "Naïve"?**

-->

The Naïve Bayes classifier is a probabilistic machine learning algorithm used for classification, based on Bayes' Theorem. [cite_start]It is termed "Naïve" because it makes a strong, often unrealistic, assumption that all features are conditionally independent of each other, given the class label[cite: 12].

**13. What is Bayes' Theorem?**

-->

Bayes' Theorem describes the probability of an event, given prior knowledge of conditions that might be related to that event. It is mathematically expressed as:
$P(A|B) = \frac{P(B|A) * P(A)}{P(B)}$
where:
* $P(A|B)$ is the posterior probability of event A given event B.
* $P(B|A)$ is the likelihood of event B given event A.
* $P(A)$ is the prior probability of event A.
* [cite_start]$P(B)$ is the prior probability of event B[cite: 13].

**14. Explain the differences between Gaussian Naïve Bayes, Multinomial Naïve Bayes, and Bernoulli Naïve Bayes.**

-->

[cite_start]These are different variants of Naïve Bayes, distinguished by the assumed probability distribution of their features[cite: 14]:
* [cite_start]**Gaussian Naïve Bayes:** This variant assumes that continuous features follow a normal (Gaussian) distribution within each class[cite: 14].
* **Multinomial Naïve Bayes:** Best suited for discrete features, such as word counts in text classification. [cite_start]It assumes a multinomial distribution for the features[cite: 14].
* **Bernoulli Naïve Bayes:** Designed for binary or boolean features (i.e., presence or absence of a feature). [cite_start]It assumes a Bernoulli distribution for these features[cite: 14].

**15. When should you use Gaussian Naïve Bayes over other variants?**

-->

[cite_start]You should use Gaussian Naïve Bayes when your features are continuous and can be reasonably assumed to follow a Gaussian (normal) distribution within each class[cite: 15].

**16. What are the key assumptions made by Naïve Bayes?**

-->

The primary and most significant assumption made by Naïve Bayes is the **conditional independence of features**. [cite_start]This means it assumes that the presence or absence of one feature does not influence the presence or absence of any other feature, given the class variable[cite: 16].

**17. What are the advantages and disadvantages of Naïve Bayes?**

-->

**Advantages:**
* [cite_start]It is simple and fast to train, even with large datasets[cite: 17].
* [cite_start]It often performs well in various real-world scenarios, particularly in text classification[cite: 17].
* [cite_start]It generally requires less training data compared to more complex models[cite: 17].
* [cite_start]It handles high-dimensional data effectively[cite: 17].

**Disadvantages:**
* [cite_start]The "naïve" assumption of feature independence rarely holds true in real-world data, which can sometimes lead to suboptimal performance[cite: 17].
* [cite_start]It can suffer from the "zero-frequency problem," where if a category is not present in the training data, the model might assign zero probability, making predictions difficult[cite: 17].
* [cite_start]It is generally not a good choice for regression tasks[cite: 17].

**18. Why is Naïve Bayes a good choice for text classification?**

-->

[cite_start]Naïve Bayes is a strong choice for text classification due to several factors[cite: 18]:
* [cite_start]It effectively handles high-dimensional and sparse data, which is common in text analysis where the vocabulary can be vast[cite: 18].
* [cite_start]Its simplicity and speed make it very efficient for processing large text datasets[cite: 18].
* [cite_start]Its probabilistic nature aligns well with the concept of word frequencies in documents[cite: 18].
* [cite_start]Despite the independence assumption, it often performs surprisingly well in text classification because the presence of one word is frequently conditionally independent of another, given the document's category[cite: 18].

**20. How does Laplace Smoothing help in Naïve Bayes?**

-->

[cite_start]Laplace Smoothing, also known as add-one smoothing, is a technique used in Naïve Bayes to address the "zero-frequency problem"[cite: 20]. This problem occurs when a specific feature value or a combination of a feature and a class label does not appear in the training data, leading to a calculated probability of zero for that combination. This can make the overall class probability zero or undefined, halting the prediction process. Laplace smoothing mitigates this by adding a small constant (typically 1) to the counts of all features and classes, ensuring that no probability ever becomes exactly zero. [cite_start]This prevents issues with unseen data combinations and makes the model more robust[cite: 20].

22. Write a Python program to train two SVM classifiers with Linear and RBF kernels on the Wine dataset, then compare their accuracies.

24. Write a Python program to train an SVM Classifier with a Polynomial Kernel and visualize the decision boundary.

```python
# Question 24: SVM with Polynomial Kernel and Decision Boundary Visualization
print("--- Question 24: SVM with Polynomial Kernel and Decision Boundary Visualization ---")

# For visualization, it's easier to use a 2D dataset
from sklearn.datasets import make_moons

X_poly, y_poly = make_moons(n_samples=100, noise=0.15, random_state=42)

# Split data
X_train_poly, X_test_poly, y_train_poly, y_test_poly = train_test_split(
    X_poly, y_poly, test_size=0.3, random_state=42
)

# Train SVM with Polynomial Kernel
svm_poly = SVC(kernel='poly', degree=3, gamma='auto', C=1, random_state=42)
svm_poly.fit(X_train_poly, y_train_poly)

# Create a meshgrid to plot decision boundary
x_min, x_max = X_poly[:, 0].min() - 1, X_poly[:, 0].max() + 1
y_min, y_max = X_poly[:, 1].min() - 1, X_poly[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                     np.arange(y_min, y_max, 0.02))

# Predict on the meshgrid
Z = svm_poly.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot the decision boundary
plt.figure(figsize=(10, 7))
plt.contourf(xx, yy, Z, alpha=0.8, cmap=plt.cm.coolwarm)
plt.scatter(X_poly[:, 0], X_poly[:, 1], c=y_poly, s=50, edgecolors='k', cmap=plt.cm.coolwarm)
plt.title('SVM with Polynomial Kernel Decision Boundary')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.colorbar(label='Class')
plt.show()

print(f"Accuracy of SVM with Polynomial Kernel: {svm_poly.score(X_test_poly, y_test_poly):.4f}")
print("-" * 70)
```

29. Write a Python program to apply feature scaling before training an SVM model and compare results with unscaled data.

```python
# Question 29: Feature Scaling with SVM
print("--- Question 29: Feature Scaling with SVM ---")

# Load the Breast Cancer dataset (features have different scales)
cancer = load_breast_cancer()
X = cancer.data
y = cancer.target

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# --- 1. Train SVM without scaling ---
svm_unscaled = SVC(kernel='rbf', gamma='auto', random_state=42) # RBF is sensitive to scaling
svm_unscaled.fit(X_train, y_train)
y_pred_unscaled = svm_unscaled.predict(X_test)
accuracy_unscaled = accuracy_score(y_test, y_pred_unscaled)
print(f"Accuracy of SVM (unscaled data): {accuracy_unscaled:.4f}")

# --- 2. Train SVM with scaling (StandardScaler) ---
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

svm_scaled = SVC(kernel='rbf', gamma='auto', random_state=42)
svm_scaled.fit(X_train_scaled, y_train)
y_pred_scaled = svm_scaled.predict(X_test_scaled)
accuracy_scaled = accuracy_score(y_test, y_pred_scaled)
print(f"Accuracy of SVM (scaled data): {accuracy_scaled:.4f}")

if accuracy_scaled > accuracy_unscaled:
    print("Feature scaling improved accuracy for SVM.")
elif accuracy_unscaled > accuracy_scaled:
    print("Unscaled data performed better (unlikely for RBF SVM, check parameters).")
else:
    print("Feature scaling had no significant impact on accuracy.")
print("-" * 70)
```

38. Write a Python program to train an SVM Classifier using Stratified K-Fold Cross-Validation and compute the average accuracy.

```python
# Question 38: SVM with Stratified K-Fold Cross-Validation
print("--- Question 38: SVM with Stratified K-Fold Cross-Validation ---")

# Load Breast Cancer dataset
cancer = load_breast_cancer()
X = cancer.data
y = cancer.target

# It's good practice to scale data before SVM
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Initialize StratifiedKFold
# n_splits is the number of folds (e.g., 5 or 10)
# shuffle=True for randomizing data before splitting
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

svm_model = SVC(kernel='rbf', gamma='scale', random_state=42) # Using RBF for this example

accuracies = []

# Perform cross-validation
for fold, (train_index, test_index) in enumerate(skf.split(X_scaled, y)):
    X_train, X_test = X_scaled[train_index], X_scaled[test_index]
    y_train, y_test = y[train_index], y[test_index]

    svm_model.fit(X_train, y_train)
    y_pred = svm_model.predict(X_test)
    fold_accuracy = accuracy_score(y_test, y_pred)
    accuracies.append(fold_accuracy)
    print(f"Fold {fold+1} Accuracy: {fold_accuracy:.4f}")

average_accuracy = np.mean(accuracies)
print(f"\nAverage accuracy across all folds: {average_accuracy:.4f}")
print("-" * 70)
```

46. Write a Python program to train an SVM Classifier and visualize the Precision-Recall Curve.

```python
# Question 46: SVM and Precision-Recall Curve Visualization
print("--- Question 46: SVM and Precision-Recall Curve Visualization ---")

# Load Breast Cancer dataset (binary classification)
cancer = load_breast_cancer()
X = cancer.data
y = cancer.target

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Scale data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Train SVM Classifier
# For precision-recall curve, we need decision function scores or probabilities.
# Set probability=True to enable predict_proba, though decision_function is often preferred for SVM.
svm_pr_curve = SVC(kernel='rbf', gamma='scale', probability=True, random_state=42)
svm_pr_curve.fit(X_train_scaled, y_train)

# Get decision function scores (preferred for SVM PR curve)
# These are not probabilities but are related to the distance from the hyperplane.
y_scores = svm_pr_curve.decision_function(X_test_scaled)

# Compute Precision-Recall Curve
precision, recall, _ = precision_recall_curve(y_test, y_scores)

# Plot Precision-Recall Curve
plt.figure(figsize=(8, 6))
plt.plot(recall, precision, color='b', alpha=0.9, lw=2)
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve for SVM Classifier')
plt.ylim([0.0, 1.05])
plt.xlim([0.0, 1.0])
plt.grid(True)
plt.show()
print("-" * 70)
```
