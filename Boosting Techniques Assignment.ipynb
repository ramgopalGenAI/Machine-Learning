1.  **What is Boosting in Machine Learning?**

    -->
    [cite_start]Boosting is a machine learning ensemble meta-algorithm primarily used to reduce bias and variance[cite: 1]. [cite_start]It works by combining multiple "weak" learners (models that are only slightly better than random guessing) to create a "strong" learner[cite: 1]. [cite_start]The core idea is to sequentially train models, with each new model focusing on correcting the errors made by the previous ones[cite: 1].

2.  **How does Boosting differ from Bagging?**

    -->
    Boosting differs from Bagging in several key ways:
    * [cite_start]**Sequential vs. Parallel:** Boosting trains models sequentially, where each new model tries to correct errors of the previous ones[cite: 1]. Bagging trains models in parallel, independently of each other.
    * **Weighting:** In Boosting, misclassified instances are given higher weights so that subsequent models focus more on them. Bagging typically gives equal weight to all instances.
    * [cite_start]**Focus:** Boosting aims to reduce bias and convert weak learners into strong ones[cite: 1]. Bagging primarily aims to reduce variance.
    * **Example:** AdaBoost and Gradient Boosting are examples of boosting, while Random Forest is an example of bagging.

3.  **What is the key idea behind AdaBoost?**

    -->
    [cite_start]The key idea behind AdaBoost (Adaptive Boosting) is to sequentially train multiple weak learners (typically decision stumps) and combine their predictions[cite: 3]. [cite_start]Each subsequent weak learner is trained to focus on the samples that were misclassified by the previous learners[cite: 3]. [cite_start]This is achieved by increasing the "weights" of the misclassified samples and decreasing the "weights" of the correctly classified samples[cite: 3]. [cite_start]Finally, the predictions of all weak learners are combined through a weighted majority vote (for classification) or weighted sum (for regression), where the weights of the weak learners depend on their accuracy[cite: 3].

4.  **Explain the working of AdaBoost with an example.**

    -->
    AdaBoost works iteratively, adjusting the focus on samples based on previous predictions. Here's a simplified explanation:

  * **Initialization:** Initially, all training samples are assigned equal weights.
  * **Iteration 1:** A weak learner (e.g., a decision stump) is trained on the dataset. The error of this learner is calculated. Samples that are misclassified are given higher weights. The learner itself is assigned a "say" or weight based on its accuracy (more accurate learners get higher say).
  * **Iteration 2 to N:** In subsequent iterations, another weak learner is trained. This new learner pays more attention to the samples that were misclassified in the previous iteration due to their increased weights. Again, its error is calculated, and its "say" is determined. The sample weights are updated based on the new learner's performance.
  * **Final Prediction:** The final prediction is a weighted sum (for regression) or weighted majority vote (for classification) of all the individual weak learners' predictions, where the weights are their respective "say" values.

  **Example:** Imagine we want to classify emails as spam or not spam.
    1.  Initially, all emails have equal importance.
    2.  A weak classifier (e.g., "if email contains 'free money', then spam") is trained. It misclassifies some emails.
    3.  The misclassified emails (e.g., legitimate emails containing "free money" in a different context) are given higher weights.
    4.  A second weak classifier (e.g., "if email sender is suspicious, then spam") is trained, paying more attention to those previously misclassified emails.
    5.  This process continues, with each new classifier trying to correct the mistakes of the previous ones.
    6.  Finally, to classify a new email, all the weak classifiers vote, and their votes are weighted by their individual accuracy.

5.  **What is Gradient Boosting, and how is it different from AdaBoost?**

    -->

   [cite_start]Gradient Boosting is a powerful boosting technique where new models are trained to predict the *residuals* or errors of the previous models, rather than directly adjusting sample weights like AdaBoost[cite: 4]. [cite_start]It builds an ensemble of weak learners (typically decision trees, often shallow ones) sequentially[cite: 4].

   [cite_start]Key differences from AdaBoost[cite: 4]:
   * **Error Correction Mechanism:** AdaBoost adjusts sample weights based on misclassifications. Gradient Boosting fits new learners to the *residuals* (the differences between the actual target values and the predictions of the current ensemble).
   * **Loss Function:** Gradient Boosting is more flexible as it can optimize arbitrary differentiable loss functions. AdaBoost is primarily designed for classification problems with an exponential loss function.
   * **Weak Learner Weights:** In AdaBoost, each weak learner is assigned a weight based on its accuracy. In Gradient Boosting, each new tree directly predicts the negative gradient of the loss function with respect to the current ensemble's prediction.
   * **Generality:** Gradient Boosting is more generalized and can be applied to a wider range of loss functions and tasks (regression, classification, ranking) compared to AdaBoost.

6.  **What is the loss function in Gradient Boosting?**

    -->
    [cite_start]In Gradient Boosting, the loss function is a differentiable function that quantifies the error between the model's predictions and the true values[cite: 5]. [cite_start]The goal of Gradient Boosting is to minimize this loss function[cite: 5]. The choice of the loss function depends on the type of problem:
    [cite_start]**Regression:** Common loss functions include Mean Squared Error (MSE), Mean Absolute Error (MAE), or Huber loss[cite: 5].
    * [cite_start]**Classification:** Common loss functions include Log-loss (for binary and multi-class classification) or exponential loss[cite: 5].

    [cite_start]During each iteration, a new weak learner is trained to predict the negative gradient of the chosen loss function with respect to the current predictions of the ensemble[cite: 5]. This essentially means that the new learner tries to "steer" the overall model towards minimizing the error by moving in the direction of the steepest descent of the loss function.

7.  **How does XGBoost improve over traditional Gradient Boosting?**

    -->
    [cite_start]XGBoost (eXtreme Gradient Boosting) is an optimized and highly efficient implementation of gradient boosting that includes several improvements over traditional Gradient Boosting[cite: 7]:
    * [cite_start]**Regularization:** XGBoost incorporates L1 (Lasso) and L2 (Ridge) regularization terms into its objective function to prevent overfitting[cite: 7]. This helps to control the complexity of the model and make it more generalized.
    * [cite_start]**Handling Missing Values:** XGBoost has a built-in mechanism to handle missing values by learning the best direction to go when a value is missing[cite: 7].
    * [cite_start]**Parallel Processing:** While individual trees are built sequentially, XGBoost can parallelize the tree construction process by using all available CPU cores during splitting to find the best split points[cite: 7].
    * [cite_start]**Tree Pruning:** XGBoost uses a 'max_depth' parameter and a 'min_child_weight' parameter to control tree growth and prunes trees backward (from leaves to root) once they are built, which helps in preventing overfitting[cite: 7].
    * [cite_start]**Smarter Splitting:** XGBoost employs an approximate greedy algorithm for splitting when exact greedy algorithms are not feasible, which can speed up computation on large datasets[cite: 7].
    * [cite_start]**Customizable Objective Function and Evaluation Metric:** Users can define their own objective functions and evaluation metrics, providing greater flexibility[cite: 7].
    * [cite_start]**Cache Awareness:** XGBoost is designed to be cache-aware, which helps in efficient data access[cite: 7].

8.  **What is the difference between XGBoost and CatBoost?**

    -->
    [cite_start]XGBoost and CatBoost are both powerful gradient boosting libraries, but they have key differences, especially in their handling of categorical features and training speed[cite: 6]:
    * **Categorical Feature Handling:**
        * **CatBoost:** Designed specifically to handle categorical features efficiently and effectively. [cite_start]It uses a novel "ordered boosting" scheme and implements a "permutation-driven" approach to transform categorical features into numerical ones, which helps avoid prediction shift and target leakage[cite: 6].
        * **XGBoost:** Requires categorical features to be pre-processed (e.g., one-hot encoding, label encoding) before training. It doesn't have built-in sophisticated handling for them.
    * **Prediction Shift and Target Leakage:**
        * [cite_start]**CatBoost:** Addresses the prediction shift problem (where the distribution of categorical features in the training data differs from the test data) and target leakage through its ordered boosting and permutation methods[cite: 6].
        * **XGBoost:** Does not inherently address these issues for categorical features; they rely on the user's pre-processing.
    * **Gradient Calculation:**
        * **CatBoost:** Uses ordered boosting, where the gradients are calculated based on a random permutation of the data to reduce bias.
        * **XGBoost:** Calculates gradients in a more traditional manner.
    * **Training Speed:**
        * **XGBoost:** Generally very fast, especially when numerical features dominate.
        * **CatBoost:** Can be slower than XGBoost on datasets with many numerical features due to its more complex categorical feature handling, but often faster and more accurate on datasets with a high proportion of categorical features.
    * **GPU Support:** Both offer GPU support for faster training.

10. **How does regularization help in XGBoost?**

    -->
    [cite_start]Regularization in XGBoost helps to prevent overfitting and improve the generalization ability of the model by controlling its complexity[cite: 7]. XGBoost incorporates two types of regularization terms into its objective function:
    * **L1 Regularization (Lasso Regression):** Adds a penalty equal to the absolute value of the magnitude of coefficients. [cite_start]This encourages sparsity in the feature weights, effectively performing feature selection by driving some coefficients to zero[cite: 7].
    * **L2 Regularization (Ridge Regression):** Adds a penalty equal to the square of the magnitude of coefficients. [cite_start]This encourages smaller coefficients and helps in preventing large individual tree weights from dominating the prediction[cite: 7].

    By adding these regularization terms, the objective function becomes:
    [cite_start]$Objective = Training Loss + Regularization Term$ [cite: 7]

    [cite_start]This means that during training, the algorithm not only tries to minimize the training error but also tries to keep the model's complexity (e.g., the number of leaves in the trees, the magnitude of the leaf weights) in check[cite: 7]. [cite_start]This leads to models that are less likely to overfit to the training data and perform better on unseen data[cite: 7].

11. **What are some hyperparameters to tune in Gradient Boosting models?**

    -->

    Gradient Boosting models have several important hyperparameters that need to be tuned to optimize performance and prevent overfitting. Some key ones include:

    * **`n_estimators` (or `num_round` in XGBoost):** The number of boosting stages or the number of weak learners (trees) to train. A higher number can lead to better performance but also increases the risk of overfitting and computation time.
    * **`learning_rate` (or `eta` in XGBoost):** Also known as shrinkage, this parameter scales the contribution of each tree. A smaller learning rate requires more estimators but makes the model more robust to overfitting.
    * **`max_depth`:** The maximum depth of each individual tree. Deeper trees can capture more complex relationships but are more prone to overfitting.
    * **`subsample`:** The fraction of samples to be used for fitting the individual base learners. Values less than 1.0 introduce randomness and can reduce variance.
    * **`min_samples_leaf` (or `min_child_weight` in XGBoost):** The minimum number of samples required to be at a leaf node. This helps in controlling overfitting by preventing the creation of leaves that are too specific to the training data.
    * **`max_features`:** The number of features to consider when looking for the best split. This can introduce randomness and help in reducing correlation between trees.
    * **`loss`:** The loss function to optimize. The choice depends on the problem (e.g., 'deviance' for classification, 'squared_error' for regression).
    * **`gamma` (XGBoost specific):** Minimum loss reduction required to make a further partition on a leaf node of the tree. Larger gamma means more conservative algorithm.
    * **`colsample_bytree` (XGBoost specific):** The subsample ratio of columns when constructing each tree.
    * **`reg_alpha` (L1 regularization) and `reg_lambda` (L2 regularization) (XGBoost specific):** Regularization terms that control the complexity of the model and prevent overfitting.
    * **`random_state`:** A seed for the random number generator, ensuring reproducibility of results.

12. **What is the concept of Feature Importance in Boosting?**

    -->
    [cite_start]Feature importance in boosting models refers to a score that indicates how useful or valuable each feature was in the construction of the boosted trees within the model[cite: 8]. [cite_start]It essentially quantifies the contribution of each feature to the overall prediction[cite: 8].

    Different boosting algorithms calculate feature importance in slightly different ways, but common methods include:
    * [cite_start]**Gain (or F-score):** This is the most common measure, especially in tree-based boosting algorithms like XGBoost and LightGBM[cite: 8]. [cite_start]It represents the total gain (improvement in purity or reduction in loss) brought by that feature across all splits in all trees[cite: 8]. A higher gain value means the feature is more important.
    * **Split Count:** This measures how many times a feature is used in a split across all trees.
    * **Weight:** Similar to split count, but it might sum the number of times a feature appears in splits weighted by the number of instances passing through those splits.

    **Why is it important?**
    * [cite_start]**Interpretability:** Feature importance helps in understanding which factors are most influential in the model's predictions, providing insights into the underlying data and problem[cite: 8].
    * [cite_start]**Feature Selection:** It can be used for feature selection, identifying and potentially removing less important features to simplify the model, reduce noise, and improve performance or training time[cite: 8].
    * **Debugging:** Understanding feature importance can help in identifying potential data quality issues or unexpected relationships in the data.
    * **Business Insights:** For business stakeholders, knowing which features drive predictions can inform strategic decisions.

13. **Why is CatBoost efficient for categorical data?**

    -->
    [cite_start]CatBoost is specifically designed to handle categorical features efficiently and effectively, which is a major reason for its strong performance on datasets with a high number of categorical variables[cite: 9]. It achieves this efficiency through several innovative techniques:

    * [cite_start]**Ordered Target Encoding (Permutation-driven):** Unlike traditional methods (like one-hot encoding or simple label encoding) that can lead to prediction shift and target leakage, CatBoost uses a sophisticated, statistics-based approach called "ordered target encoding"[cite: 9]. It transforms categorical features into numerical ones on the fly during training. [cite_start]For each sample, the categorical feature value is converted into a numerical value based on the average target value of *previous* samples with the same category, ordered by a random permutation[cite: 9]. This prevents information from the current sample's target from leaking into its feature transformation, thus avoiding target leakage.
    * [cite_start]**No Need for One-Hot Encoding:** CatBoost automatically handles categorical features without requiring manual one-hot encoding or other pre-processing steps, which simplifies the data pipeline and can save memory and computation for high-cardinality categorical features[cite: 9].
    * [cite_start]**Oblivious Trees:** CatBoost uses "oblivious trees" as its base learners[cite: 9]. In an oblivious tree, the same splitting criterion is used for all nodes at the same level of the tree. [cite_start]This helps in regularizing the model, preventing overfitting, and allowing for faster prediction, as the tree structure is symmetric[cite: 9].
    * [cite_start]**Ordered Boosting:** CatBoost uses a variant of gradient boosting called "ordered boosting"[cite: 9]. This method reduces prediction shift, which is a problem where the distribution of categorical features in the training data differs from the test data. [cite_start]By ordering samples randomly and training on subsets, it creates more robust and unbiased models[cite: 9].

    [cite_start]These internal mechanisms allow CatBoost to effectively capture complex relationships within categorical data while mitigating common issues like target leakage and prediction shift, making it a highly efficient and accurate choice for datasets rich in categorical features[cite: 9].

14. Train an AdaBoost Classifier on a sample dataset and print model accuracy.

```python
import numpy as np
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# 1. Generate a sample dataset
X, y = make_classification(n_samples=1000, n_features=10, n_informative=5, n_redundant=0, random_state=42)

# 2. Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 3. Initialize the AdaBoost Classifier
# A common weak learner is a Decision Tree Classifier (often a shallow one, like max_depth=1 for a Decision Stump)
base_estimator = DecisionTreeClassifier(max_depth=1)
ada_clf = AdaBoostClassifier(estimator=base_estimator, n_estimators=100, random_state=42)

# 4. Train the AdaBoost Classifier
ada_clf.fit(X_train, y_train)

# 5. Make predictions on the test set
y_pred = ada_clf.predict(X_test)

# 6. Calculate and print the model accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"AdaBoost Classifier Accuracy: {accuracy:.4f}")
```

18. Train an XGBoost Classifier on a dataset and compare accuracy with Gradient Boosting.

```python
import numpy as np
from sklearn.ensemble import GradientBoostingClassifier
from xgboost import XGBClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# 1. Generate a sample classification dataset
X, y = make_classification(n_samples=1000, n_features=10, n_informative=5, n_redundant=0, random_state=42)

# 2. Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 3. Train a Gradient Boosting Classifier
gb_clf = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)
gb_clf.fit(X_train, y_train)
gb_pred = gb_clf.predict(X_test)
gb_accuracy = accuracy_score(y_test, gb_pred)
print(f"Gradient Boosting Classifier Accuracy: {gb_accuracy:.4f}")

# 4. Train an XGBoost Classifier
xgb_clf = XGBClassifier(n_estimators=100, learning_rate=0.1, max_depth=3, use_label_encoder=False, eval_metric='logloss', random_state=42)
xgb_clf.fit(X_train, y_train)
xgb_pred = xgb_clf.predict(X_test)
xgb_accuracy = accuracy_score(y_test, xgb_pred)
print(f"XGBoost Classifier Accuracy: {xgb_accuracy:.4f}")

# 5. Compare accuracies
if xgb_accuracy > gb_accuracy:
    print("\nXGBoost Classifier performed better.")
elif gb_accuracy > xgb_accuracy:
    print("\nGradient Boosting Classifier performed better.")
else:
    print("\nBoth classifiers performed equally well.")
```

22. Train a Gradient Boosting Regressor and plot learning curves.

```python
import numpy as np
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

# 1. Generate a sample regression dataset
X, y = make_regression(n_samples=500, n_features=5, n_informative=3, random_state=42)

# 2. Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 3. Initialize the Gradient Boosting Regressor
gb_reg = GradientBoostingRegressor(n_estimators=200, learning_rate=0.1, max_depth=3, random_state=42)

# 4. Train the Gradient Boosting Regressor and record errors at each stage
# Note: scikit-learn's GradientBoosting models provide `staged_predict`
# or we can manually iterate to plot learning curves effectively.
# For learning curve, we typically plot error vs number of estimators

train_errors = []
test_errors = []

for i in range(1, gb_reg.n_estimators + 1):
    gb_reg.set_params(n_estimators=i) # Set number of estimators for current iteration
    gb_reg.fit(X_train, y_train)

    y_train_pred = gb_reg.predict(X_train)
    y_test_pred = gb_reg.predict(X_test)

    train_errors.append(mean_squared_error(y_train, y_train_pred))
    test_errors.append(mean_squared_error(y_test, y_test_pred))

# 5. Plot the learning curves
plt.figure(figsize=(10, 6))
plt.plot(np.arange(1, gb_reg.n_estimators + 1), train_errors, label='Training MSE')
plt.plot(np.arange(1, gb_reg.n_estimators + 1), test_errors, label='Testing MSE')
plt.xlabel("Number of Estimators")
plt.ylabel("Mean Squared Error (MSE)")
plt.title("Gradient Boosting Regressor Learning Curves")
plt.legend()
plt.grid(True)
plt.show()
```

27. Train an XGBoost Regressor and tune the learning rate using GridSearchCV.

```python
import numpy as np
from xgboost import XGBRegressor
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import mean_squared_error
import warnings

# Suppress warnings that might arise from GridSearchCV
warnings.filterwarnings('ignore')

# 1. Generate a sample regression dataset
X, y = make_regression(n_samples=500, n_features=5, n_informative=3, random_state=42)

# 2. Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 3. Define the parameter grid for learning_rate
param_grid = {
    'learning_rate': [0.01, 0.05, 0.1, 0.2]
}

# 4. Initialize the XGBoost Regressor
xgb_reg = XGBRegressor(n_estimators=100, max_depth=3, random_state=42)

# 5. Initialize GridSearchCV
# Using neg_mean_squared_error because GridSearchCV minimizes the score, so we negate MSE
grid_search = GridSearchCV(estimator=xgb_reg, param_grid=param_grid,
                           scoring='neg_mean_squared_error', cv=3, verbose=1, n_jobs=-1)

# 6. Perform GridSearchCV to find the best learning rate
grid_search.fit(X_train, y_train)

# 7. Print the best parameters and best score
print(f"Best parameters found: {grid_search.best_params_}")
print(f"Best negative MSE (training): {grid_search.best_score_:.4f}")

# 8. Evaluate the best model on the test set
best_xgb_reg = grid_search.best_estimator_
y_pred = best_xgb_reg.predict(X_test)
test_mse = mean_squared_error(y_test, y_pred)
print(f"Test MSE with best learning rate: {test_mse:.4f}")
```

```python
# ==============================================================================
# Complete Python Code for Boosting Techniques Practical Questions (14-30)
# Designed for use in Google Colaboratory (Colab)
# ==============================================================================

# --- Step 1: Install necessary libraries (run this cell first in Colab) ---
# If you run into "No module named 'catboost'" or similar errors,
# uncomment the lines below and run this cell.
# !pip install catboost
# !pip install xgboost
# !pip install scikit-learn
# !pip install matplotlib
# !pip install seaborn
# !pip install pandas
print("Attempting to install CatBoost, XGBoost, and other required libraries...")
try:
    import catboost
    print("CatBoost already installed.")
except ImportError:
    print("Installing CatBoost...")
    !pip install catboost
    print("CatBoost installed.")

try:
    import xgboost
    print("XGBoost already installed.")
except ImportError:
    print("Installing XGBoost...")
    !pip install xgboost
    print("XGBoost installed.")

# Standard libraries check (usually pre-installed in Colab)
try:
    import sklearn
    import matplotlib
    import seaborn
    import pandas
    print("Scikit-learn, Matplotlib, Seaborn, Pandas are installed.")
except ImportError:
    print("Some standard libraries might be missing. Proceeding with installation for all common ML libs...")
    !pip install scikit-learn matplotlib seaborn pandas
    print("Standard ML libraries installed.")

print("\n--- Step 2: Import all necessary libraries ---")
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Scikit-learn imports for various models, datasets, and metrics
from sklearn.ensemble import AdaBoostClassifier, AdaBoostRegressor
from sklearn.ensemble import GradientBoostingClassifier, GradientBoostingRegressor
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
from sklearn.datasets import (
    make_classification,
    make_regression,
    load_breast_cancer,
)
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import (
    accuracy_score,
    mean_absolute_error,
    r2_score,
    f1_score,
    confusion_matrix,
    ConfusionMatrixDisplay,
    roc_curve,
    auc,
    log_loss,
    mean_squared_error,
    classification_report,
)

# XGBoost import (using the common alias 'xgb')
import xgboost as xgb

# CatBoost import
from catboost import CatBoostClassifier, CatBoostRegressor

# For handling imbalanced dataset class distribution
from collections import Counter

# To suppress warnings that might arise during execution (e.g., from GridSearchCV)
import warnings
warnings.filterwarnings('ignore')

print("All necessary libraries imported successfully.")
print("Proceeding with the practical questions.\n")
# ==============================================================================
# Q28. Train a CatBoost Classifier on an imbalanced dataset and compare
#      performance with class weighting.
# ==============================================================================
print("="*80)
print("Q28. Train a CatBoost Classifier on an imbalanced dataset and compare performance with class weighting.")
print("="*80)
# 1. Generate an imbalanced sample dataset
X_q28, y_q28 = make_classification(n_samples=1000, n_features=10, n_informative=5, n_redundant=0,
                           weights=[0.9, 0.1], # 90% majority class (0), 10% minority class (1)
                           flip_y=0, # Noisy labels for demonstration purposes (set to 0 for cleaner imbalance)
                           random_state=42)

print(f"Original dataset class distribution: {Counter(y_q28)}")

# 2. Split the dataset into training and testing sets
# Use 'stratify=y' to ensure that both training and testing sets have a
# similar class distribution as the original dataset. This is crucial for
# imbalanced datasets.
X_train_q28, X_test_q28, y_train_q28, y_test_q28 = train_test_split(X_q28, y_q28, test_size=0.3, random_state=42, stratify=y_q28)
print(f"Training set class distribution: {Counter(y_train_q28)}")
print(f"Test set class distribution: {Counter(y_test_q28)}")

# --- Model 1: CatBoost without class weighting ---
print("\n--- CatBoost Classifier (No Class Weighting) ---")
cat_clf_no_weight_q28 = CatBoostClassifier(iterations=100, learning_rate=0.1, depth=3,
                                       random_seed=42, verbose=0) # verbose=0 to suppress training output
cat_clf_no_weight_q28.fit(X_train_q28, y_train_q28)
y_pred_no_weight_q28 = cat_clf_no_weight_q28.predict(X_test_q28)

print("Classification Report (No Weighting):\n", classification_report(y_test_q28, y_pred_no_weight_q28))
print(f"Accuracy (No Weighting): {accuracy_score(y_test_q28, y_pred_no_weight_q28):.4f}")
print(f"F1-Score (No Weighting): {f1_score(y_test_q28, y_pred_no_weight_q28):.4f}")


# --- Model 2: CatBoost with class weighting ---
print("\n--- CatBoost Classifier (With Class Weighting) ---")

# Calculate class weights: This is a common strategy to give more importance
# to the minority class. A common method is inverse proportionality to class frequency.
# The formula `total_samples / (num_classes * count_of_class)` is often used.
class_counts_q28 = Counter(y_train_q28)
total_samples_q28 = sum(class_counts_q28.values())

# For binary classification (0 and 1)
# weight for class 0 = total_samples / (2 * count of class 0)
# weight for class 1 = total_samples / (2 * count of class 1)
class_weights_q28 = {
    0: total_samples_q28 / (2 * class_counts_q28[0]),
    1: total_samples_q28 / (2 * class_counts_q28[1])
}
print(f"Calculated Class Weights: {class_weights_q28}")

cat_clf_with_weight_q28 = CatBoostClassifier(iterations=100, learning_rate=0.1, depth=3,
                                         random_seed=42, verbose=0,
                                         class_weights=class_weights_q28) # Apply class weights
cat_clf_with_weight_q28.fit(X_train_q28, y_train_q28)
y_pred_with_weight_q28 = cat_clf_with_weight_q28.predict(X_test_q28)

print("Classification Report (With Weighting):\n", classification_report(y_test_q28, y_pred_with_weight_q28))
print(f"Accuracy (With Weighting): {accuracy_score(y_test_q28, y_pred_with_weight_q28):.4f}")
print(f"F1-Score (With Weighting): {f1_score(y_test_q28, y_pred_with_weight_q28):.4f}")

print("\n--- Comparison (Q28) ---")
print("Observe F1-score and recall for the minority class (class 1) in the classification reports.")
print("With class weighting, the model should ideally show improved performance metrics (like F1-score and recall) for the minority class, even if overall accuracy drops slightly, as it prioritizes correct prediction of the rarer class.\n")
```
