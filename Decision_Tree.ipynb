1. What is a Decision Tree, and how does it work?

A Decision Tree is a supervised learning algorithm for classification and regression. It works by recursively splitting data into subsets based on significant features, forming a tree. Internal nodes test attributes, branches represent outcomes, and leaf nodes provide predictions (class labels or numerical values). The process traverses from the root to a leaf for a prediction.

2. What are impurity measures in Decision Trees?

Impurity measures quantify the homogeneity of samples. Decision Trees aim to find splits that maximize impurity reduction. Common measures include Gini Impurity and Entropy; a pure node contains samples of a single class.

6. What is the difference between Gini Impurity and Entropy?

Both measure impurity for splits. Gini Impurity tends to isolate the most frequent class, while Entropy aims for a more balanced tree. Gini is computationally less intensive. Practically, their impact on model performance is often minimal.

7. What is the mathematical explanation behind Decision Trees?

Decision Trees mathematically seek optimal splits to separate data into homogeneous groups. This involves quantifying disorder (impurity), calculating impurity reduction (information gain) to greedily select the best split, recursively partitioning subsets, and finally predicting at leaf nodes (majority class for classification, average/median for regression).

8. What is Pre-Pruning in Decision Trees?

Pre-pruning (early stopping) halts tree growth before it's fully formed, preventing overfitting. This is achieved by setting constraints like max_depth, min_samples_split, min_samples_leaf, or max_features during construction.

9. What is Post-Pruning in Decision Trees?

Post-pruning involves growing a full tree, then removing branches or nodes that don't significantly improve accuracy on a validation set. This often means replacing subtrees with leaf nodes if error rates don't increase. Cost Complexity Pruning (CCP) is a common method.

10. What is the difference between Pre-Pruning and Post-Pruning?

Pre-pruning stops tree construction early; it's faster but risks missing good splits ("top-down"). Post-pruning grows the full tree then prunes it back; it's more effective but computationally costlier ("bottom-up").

11. What is a Decision Tree Regressor?

A Decision Tree Regressor is used for predicting continuous numerical values. Instead of class labels, leaf nodes predict the average (or median) target value of samples within them. Splitting criteria aim to minimize variance or Mean Squared Error (MSE) within nodes.

12. What are the advantages and disadvantages of Decision Trees?

Advantages:

Easy to understand and interpret.

Requires minimal data preparation (e.g., no scaling).

Handles both numerical and categorical data.

Non-parametric; no data distribution assumptions.

Models non-linear relationships.

Disadvantages:

Prone to overfitting noisy data.

Unstable; minor data changes can alter tree structure.

Biased towards dominant classes with imbalanced data.

Greedy approach doesn't guarantee global optimum.

13. How does a Decision Tree handle missing values?

Decision Trees handle missing values by ignoring samples for specific features, imputing values (mean/median/mode), using surrogate splits (alternative features), or probabilistically assigning samples to paths.

14. How does a Decision Tree handle categorical features?

Decision Trees handle categorical features via binary splits (grouping categories), multi-way splits (one branch per category), or commonly, by converting them to numerical features using one-hot encoding.

```python
# -*- coding: utf-8 -*-
"""
Practical Decision Tree Implementations in Python
This Colab notebook provides solutions to the practical Decision Tree assignment questions.
Each question is addressed in a separate cell.
"""

# Import necessary libraries
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris, fetch_california_housing
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor, export_graphviz
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score, mean_squared_error, precision_score, recall_score, f1_score, confusion_matrix
from sklearn.preprocessing import StandardScaler
from sklearn.multiclass import OneVsRestClassifier
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import Image # Used for displaying graphviz output in Colab
import pydotplus # Make sure to install: pip install pydotplus graphviz

# Suppress warnings for cleaner output
import warnings
warnings.filterwarnings('ignore')

# --- Question 16: Train a Decision Tree Classifier on the Iris dataset and print the model accuracy. ---
print("--- Question 16: Decision Tree Classifier on Iris dataset ---") # [cite: 10, 34]
# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize and train the Decision Tree Classifier
dt_classifier_q16 = DecisionTreeClassifier(random_state=42)
dt_classifier_q16.fit(X_train, y_train)

# Make predictions on the test set
y_pred_q16 = dt_classifier_q16.predict(X_test)

# Calculate and print the accuracy
accuracy_q16 = accuracy_score(y_test, y_pred_q16)
print(f"Model Accuracy (Iris dataset): {accuracy_q16:.4f}\n")

# --- Question 17: Train a Decision Tree Classifier using Gini Impurity as the criterion and print the feature importances. ---
print("--- Question 17: Gini Impurity and Feature Importances ---") # [cite: 11, 35]
# Initialize and train the Decision Tree Classifier with Gini criterion (default)
dt_classifier_gini = DecisionTreeClassifier(criterion='gini', random_state=42)
dt_classifier_gini.fit(X_train, y_train)

# Print feature importances
print("Feature Importances (Gini):")
for feature, importance in zip(iris.feature_names, dt_classifier_gini.feature_importances_):
    print(f"- {feature}: {importance:.4f}")
print("\n")

# --- Question 18: Train a Decision Tree Classifier using Entropy as the splitting criterion and print the model accuracy. ---
print("--- Question 18: Entropy as Splitting Criterion ---") # [cite: 12, 36]
# Initialize and train the Decision Tree Classifier with Entropy criterion
dt_classifier_entropy = DecisionTreeClassifier(criterion='entropy', random_state=42)
dt_classifier_entropy.fit(X_train, y_train)

# Make predictions on the test set
y_pred_entropy = dt_classifier_entropy.predict(X_test)

# Calculate and print the accuracy
accuracy_entropy = accuracy_score(y_test, y_pred_entropy)
print(f"Model Accuracy (Entropy criterion): {accuracy_entropy:.4f}\n")

# --- Question 19: Train a Decision Tree Regressor on a housing dataset and evaluate using Mean Squared Error (MSE). ---
print("--- Question 19: Decision Tree Regressor on Housing dataset ---") # [cite: 13, 37]
# Load the California Housing dataset (Boston housing is deprecated)
housing = fetch_california_housing()
X_housing = housing.data
y_housing = housing.target

# Split the dataset
X_train_housing, X_test_housing, y_train_housing, y_test_housing = train_test_split(
    X_housing, y_housing, test_size=0.3, random_state=42
)

# Initialize and train the Decision Tree Regressor
dt_regressor = DecisionTreeRegressor(random_state=42)
dt_regressor.fit(X_train_housing, y_train_housing)

# Make predictions
y_pred_housing = dt_regressor.predict(X_test_housing)

# Calculate and print MSE
mse_housing = mean_squared_error(y_test_housing, y_pred_housing)
print(f"Mean Squared Error (Housing dataset): {mse_housing:.4f}\n")

# --- Question 20: Train a Decision Tree Classifier and visualize the tree using graphviz. ---
print("--- Question 20: Visualize Decision Tree with Graphviz ---") # [cite: 14, 38]
# Using the Iris dataset and the classifier from Q16
# Export the tree to a DOT file
dot_data = export_graphviz(
    dt_classifier_q16,
    out_file=None,
    feature_names=iris.feature_names,
    class_names=iris.target_names,
    filled=True,
    rounded=True,
    special_characters=True
)

# Create a graph from the DOT data
graph = pydotplus.graph_from_dot_data(dot_data)

# Save the graph as a PNG image (optional, for local viewing)
# graph.write_png("iris_decision_tree.png")

# Display the image in Colab (requires graphviz and pydotplus to be installed)
# In a real Colab environment, this would render the image directly.
# For this structured output, we'll just indicate it's generated.
print("Decision tree visualization generated. (Image would be displayed in Colab output)")
# Image(graph.create_png()) # Uncomment this line if running directly in a Colab notebook
print("\n")

# --- Question 21: Train a Decision Tree Classifier with a maximum depth of 3 and compare its accuracy with a fully grown tree. ---
print("--- Question 21: Max Depth Comparison ---") # [cite: 15, 39]
# Fully grown tree accuracy (from Q16)
accuracy_full_tree = accuracy_q16
print(f"Accuracy of fully grown tree: {accuracy_full_tree:.4f}")

# Train a Decision Tree Classifier with max_depth=3
dt_classifier_depth3 = DecisionTreeClassifier(max_depth=3, random_state=42)
dt_classifier_depth3.fit(X_train, y_train)

# Make predictions
y_pred_depth3 = dt_classifier_depth3.predict(X_test)

# Calculate and print accuracy for max_depth=3 tree
accuracy_depth3 = accuracy_score(y_test, y_pred_depth3)
print(f"Accuracy of tree with max_depth=3: {accuracy_depth3:.4f}")

if accuracy_depth3 < accuracy_full_tree:
    print("The tree with max_depth=3 has lower accuracy, indicating some underfitting or simpler model.")
elif accuracy_depth3 > accuracy_full_tree:
    print("The tree with max_depth=3 has higher accuracy, suggesting the full tree might be overfitting.")
else:
    print("The accuracies are similar.")
print("\n")

# --- Question 22: Train a Decision Tree Classifier using min_samples_split=5 and compare its accuracy with a default tree. ---
print("--- Question 22: min_samples_split Comparison ---") # [cite: 16, 40]
# Default tree accuracy (from Q16)
print(f"Accuracy of default tree: {accuracy_q16:.4f}")

# Train a Decision Tree Classifier with min_samples_split=5
dt_classifier_min_split5 = DecisionTreeClassifier(min_samples_split=5, random_state=42)
dt_classifier_min_split5.fit(X_train, y_train)

# Make predictions
y_pred_min_split5 = dt_classifier_min_split5.predict(X_test)

# Calculate and print accuracy for min_samples_split=5 tree
accuracy_min_split5 = accuracy_score(y_test, y_pred_min_split5)
print(f"Accuracy of tree with min_samples_split=5: {accuracy_min_split5:.4f}")

if accuracy_min_split5 < accuracy_q16:
    print("The tree with min_samples_split=5 has lower accuracy, indicating more aggressive pruning.")
elif accuracy_min_split5 > accuracy_q16:
    print("The tree with min_samples_split=5 has higher accuracy, suggesting it generalizes better.")
else:
    print("The accuracies are similar.")
print("\n")

# --- Question 23: Apply feature scaling before training a Decision Tree Classifier and compare its accuracy with unscaled data. ---
print("--- Question 23: Feature Scaling Impact ---") # [cite: 17, 41]
# Unscaled data accuracy (from Q16)
print(f"Accuracy with unscaled data: {accuracy_q16:.4f}")

# Apply StandardScaler to the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split the scaled dataset
X_train_scaled, X_test_scaled, y_train_scaled, y_test_scaled = train_test_split(
    X_scaled, y, test_size=0.3, random_state=42
)

# Train a Decision Tree Classifier on scaled data
dt_classifier_scaled = DecisionTreeClassifier(random_state=42)
dt_classifier_scaled.fit(X_train_scaled, y_train_scaled)

# Make predictions
y_pred_scaled = dt_classifier_scaled.predict(X_test_scaled)

# Calculate and print accuracy for scaled data
accuracy_scaled = accuracy_score(y_test_scaled, y_pred_scaled)
print(f"Accuracy with scaled data: {accuracy_scaled:.4f}")

print("Decision Trees are generally insensitive to feature scaling because their splitting decisions are based on thresholds of individual features, not distances.")
print("Therefore, scaling typically has little to no impact on their performance.\n")

# --- Question 24: Train a Decision Tree Classifier using One-vs-Rest (OvR) strategy for multiclass classification. ---
print("--- Question 24: One-vs-Rest (OvR) Strategy ---") # [cite: 18, 42]
# Initialize the base Decision Tree Classifier
base_classifier = DecisionTreeClassifier(random_state=42)

# Initialize OneVsRestClassifier with the base classifier
ovr_classifier = OneVsRestClassifier(base_classifier)

# Train the OvR classifier
ovr_classifier.fit(X_train, y_train)

# Make predictions
y_pred_ovr = ovr_classifier.predict(X_test)

# Calculate and print accuracy
accuracy_ovr = accuracy_score(y_test, y_pred_ovr)
print(f"Accuracy with One-vs-Rest strategy: {accuracy_ovr:.4f}")
print("Note: Scikit-learn's DecisionTreeClassifier inherently handles multiclass classification,")
print("so OvR might not show a significant difference in accuracy for this dataset.\n")

# --- Question 25: Train a Decision Tree Classifier and display the feature importance scores. ---
print("--- Question 25: Display Feature Importance Scores ---") # [cite: 19, 43]
# Using the classifier from Q16 (dt_classifier_q16)
print("Feature Importances:")
for feature, importance in zip(iris.feature_names, dt_classifier_q16.feature_importances_):
    print(f"- {feature}: {importance:.4f}")
print("\n")

# --- Question 26: Train a Decision Tree Regressor with max_depth=5 and compare its performance with an unrestricted tree. ---
print("--- Question 26: Regressor Max Depth Comparison ---") # [cite: 20, 44]
# Unrestricted tree MSE (from Q19)
print(f"MSE of unrestricted Regressor tree: {mse_housing:.4f}")

# Train a Decision Tree Regressor with max_depth=5
dt_regressor_depth5 = DecisionTreeRegressor(max_depth=5, random_state=42)
dt_regressor_depth5.fit(X_train_housing, y_train_housing)

# Make predictions
y_pred_regressor_depth5 = dt_regressor_depth5.predict(X_test_housing)

# Calculate and print MSE for max_depth=5 tree
mse_regressor_depth5 = mean_squared_error(y_test_housing, y_pred_regressor_depth5)
print(f"MSE of Regressor tree with max_depth=5: {mse_regressor_depth5:.4f}")

if mse_regressor_depth5 > mse_housing:
    print("The tree with max_depth=5 has higher MSE, suggesting underfitting due to restricted depth.")
elif mse_regressor_depth5 < mse_housing:
    print("The tree with max_depth=5 has lower MSE, indicating better generalization (less overfitting).")
else:
    print("The MSE values are similar.")
print("\n")

# --- Question 27: Train a Decision Tree Classifier, apply Cost Complexity Pruning (CCP), and visualize its effect on accuracy. ---
print("--- Question 27: Cost Complexity Pruning (CCP) ---") # [cite: 21, 45]
# Get alpha values for pruning
path = dt_classifier_q16.cost_complexity_pruning_path(X_train, y_train)
ccp_alphas, impurities = path.ccp_alphas, path.impurities

# Store accuracies for different alpha values
accuracies_train, accuracies_test = [], []
for ccp_alpha in ccp_alphas:
    dt_pruned = DecisionTreeClassifier(random_state=42, ccp_alpha=ccp_alpha)
    dt_pruned.fit(X_train, y_train)
    y_train_pred = dt_pruned.predict(X_train)
    y_test_pred = dt_pruned.predict(X_test)
    accuracies_train.append(accuracy_score(y_train, y_train_pred))
    accuracies_test.append(accuracy_score(y_test, y_test_pred))

# Plot the accuracy vs alpha
plt.figure(figsize=(10, 6))
plt.plot(ccp_alphas, accuracies_train, marker='o', label='train accuracy', drawstyle="steps-post")
plt.plot(ccp_alphas, accuracies_test, marker='o', label='test accuracy', drawstyle="steps-post")
plt.xlabel("alpha")
plt.ylabel("accuracy")
plt.title("Accuracy vs. alpha for training and test sets (CCP)")
plt.legend()
plt.grid(True)
plt.show()

# Find the optimal alpha that maximizes test accuracy
optimal_alpha_idx = np.argmax(accuracies_test)
optimal_ccp_alpha = ccp_alphas[optimal_alpha_idx]
print(f"Optimal ccp_alpha: {optimal_ccp_alpha:.4f}")

# Train the tree with optimal ccp_alpha
dt_optimal_pruned = DecisionTreeClassifier(random_state=42, ccp_alpha=optimal_ccp_alpha)
dt_optimal_pruned.fit(X_train, y_train)
y_pred_optimal_pruned = dt_optimal_pruned.predict(X_test)
accuracy_optimal_pruned = accuracy_score(y_test, y_pred_optimal_pruned)
print(f"Accuracy with optimal CCP: {accuracy_optimal_pruned:.4f}\n")

# --- Question 28: Train a Decision Tree Classifier and evaluate its performance using Precision, Recall, and F1-Score. ---
print("--- Question 28: Precision, Recall, F1-Score ---") # [cite: 22, 46]
# Using the classifier from Q16 (dt_classifier_q16) and its predictions (y_pred_q16)
precision = precision_score(y_test, y_pred_q16, average='weighted')
recall = recall_score(y_test, y_pred_q16, average='weighted')
f1 = f1_score(y_test, y_pred_q16, average='weighted')

print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-Score: {f1:.4f}\n")

# --- Question 29: Train a Decision Tree Classifier and visualize the confusion matrix using seaborn. ---
print("--- Question 29: Visualize Confusion Matrix ---") # [cite: 23, 47]
# Using the classifier from Q16 (dt_classifier_q16) and its predictions (y_pred_q16)
cm = confusion_matrix(y_test, y_pred_q16)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=iris.target_names, yticklabels=iris.target_names)
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix for Iris Decision Tree Classifier')
plt.show()
print("Confusion matrix displayed above.\n")

# --- Question 30: Train a Decision Tree Classifier and use GridSearchCV to find the optimal values for max_depth and min_samples_split. ---
print("--- Question 30: GridSearchCV for Hyperparameter Tuning ---") # [cite: 24, 48]
# Define the parameter grid
param_grid = {
    'max_depth': [2, 3, 4, 5, 6, 7, 8, None], # None means full depth
    'min_samples_split': [2, 5, 10, 15, 20]
}

# Initialize GridSearchCV
grid_search = GridSearchCV(
    DecisionTreeClassifier(random_state=42),
    param_grid,
    cv=5, # 5-fold cross-validation
    scoring='accuracy',
    n_jobs=-1 # Use all available cores
)

# Perform GridSearchCV
grid_search.fit(X_train, y_train)

# Print the best parameters and best score
print(f"Best parameters found: {grid_search.best_params_}")
print(f"Best cross-validation accuracy: {grid_search.best_score_:.4f}")

# Evaluate on the test set with the best estimator
best_dt_classifier = grid_search.best_estimator_
y_pred_grid = best_dt_classifier.predict(X_test)
accuracy_grid = accuracy_score(y_test, y_pred_grid)
print(f"Test set accuracy with optimal parameters: {accuracy_grid:.4f}\n")
```
